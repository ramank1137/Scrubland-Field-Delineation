```python
import requests

zenodo_url = 'https://zenodo.org/records/13952806'
samples_fn = 'gpw_grassland_fscs.vi.vhr.overlaid_point.samples_20000101_20221231_go_epsg.4326_v1.pq'
samples_url = f'{zenodo_url}/files/{samples_fn}?download=1'

samples_req = requests.get(samples_url, allow_redirects=True)

print(f"Downloading {samples_fn}")
open(samples_fn, 'wb').write(samples_req.content)
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
samples = pd.read_parquet(samples_fn)
samples
```

```python
df = samples
min_x = df['x'].min()
max_x = df['x'].max()

min_y = df['y'].min()
max_y = df['y'].max()

print("x: min =", min_x, ", max =", max_x)
print("y: min =", min_y, ", max =", max_y)
```

```python
import pandas as pd
import ee
import geemap.foliumap as geemap

# Initialize Earth Engine
ee.Initialize()



# Define color mapping for each class
color_map = {
    'Cultivated grassland': 'green',
    'Natural/Semi-natural grassland': 'orange',
    'Other land cover': 'blue'
}

# Create an empty FeatureCollection list
features = []

# Convert each row to an ee.Feature
for _, row in df.iterrows():
    point = ee.Geometry.Point([row['x'], row['y']])
    feature = ee.Feature(point, {
        'class_label': row['class_label'],
        'style': {
            'color': color_map[row['class_label']],
            'pointSize': 6,
            'pointShape': 'circle'
        }
    })
    features.append(feature)

# Create FeatureCollection
fc = ee.FeatureCollection(features)

# Create Map
Map = geemap.Map(center=[0, 0], zoom=2)
Map.addLayer(fc.style(**{
    'styleProperty': 'style'
}), {}, "Labeled Points")
Map.add_legend(title="Class Labels", labels=list(color_map.keys()), colors=list(color_map.values()))
Map
```

```python
import geopandas as gpd
# Create geometry column
gdf = gpd.GeoDataFrame(samples[["class", "class_label", "x", "y", "imagery", "ref_date", "year"]], geometry=gpd.points_from_xy(samples['x'], samples['y']), crs="EPSG:4326")

# Save to GeoJSON
gdf.to_file("wri_data/class_data.shp")
```
