```python
import requests
from pathlib import Path
import requests, os
# choose a persistent location; here: ./data under the current folder
fname = 'gpw_grassland_fscs.vi.vhr.overlaid_point.samples_20000101_20221231_go_epsg.4326_v1.pq'
save_dir = Path("../data")
save_dir.mkdir(parents=True, exist_ok=True)
save_path = save_dir / fname
zenodo_url = 'https://zenodo.org/records/13952806'
url = f'{zenodo_url}/files/{fname}?download=1'
# skip re-download if already present
if not save_path.exists():
    print(f"Downloading to: {save_path}")
    r = requests.get(url, stream=True)
    r.raise_for_status()
    with open(save_path, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
else:
    print(f"File already exists: {save_path}")
```

```python
import geemap

import re, ee
roi = ee.FeatureCollection("users/mtpictd/agro_eco_regions")
fc = fc.filterBounds(roi)
original_lulc_img = ee.Image('projects/corestack-datasets/assets/datasets/LULC_v3_river_basin/pan_india_lulc_v3_2023_2024').select('predicted_label')
wri_map = ee.ImageCollection("projects/global-pasture-watch/assets/ggc-30m/v1/grassland_c").mosaic().clip(roi)
PARENT = 'projects/raman-461708/assets'

# ---- List all image assets once (fast follow-up lookups) ----
def _basename(asset_name: str) -> str:
    return asset_name.split('/')[-1]

all_assets = ee.data.listAssets({'parent': PARENT}).get('assets', [])
image_assets = [a for a in all_assets if a.get('type','').lower() == 'image']
basenames = {_basename(a['name']) for a in image_assets}

def asset_exists_base(base: str) -> bool:
    return base in basenames

def fullpath(base: str) -> str:
    return f'{PARENT}/{base}'

def list_part_bases_for_aez(aez_no: int):
    """Return sorted basenames for all parts like AEZ_{n}_v4_local_partK (K numeric)."""
    rx = re.compile(rf'^AEZ_{aez_no}_v4_local_part(\d+)$')
    parts = []
    for b in basenames:
        m = rx.match(b)
        if m:
            parts.append((int(m.group(1)), b))
    # Sort by numeric suffix to keep deterministic order (not required for mosaic, but nice to have).
    parts.sort(key=lambda x: x[0])
    return [b for _, b in parts]

def choose_assets_for_aez(aez_no: int):
    """
    Preference:
      1) AEZ_{n}_v4
      2) AEZ_{n}_v4_local_part*
      3) AEZ_{n}_v4_local
    Returns: list of full asset IDs (possibly multiple if parts), plus a metadata dict.
    """
    base_v4 = f'AEZ_{aez_no}_v4'
    base_local = f'AEZ_{aez_no}_v4_local'
    part_bases = list_part_bases_for_aez(aez_no)

    if asset_exists_base(base_v4):
        return [fullpath(base_v4)], {'source':'v4','parts':[]}

    if part_bases:
        return [fullpath(b) for b in part_bases], {'source':'v4_local_parts','parts':part_bases}

    if asset_exists_base(base_local):
        return [fullpath(base_local)], {'source':'v4_local','parts':[]}

    return [], {'source':'missing','parts':[]}

# ---- Collect images AEZ 1..19 ----
images = []
report = {}   # aez_no -> {'source':..., 'parts':[...] }
missing = []

for aez_no in range(1, 20):
    asset_ids, meta = choose_assets_for_aez(aez_no)
    report[aez_no] = meta
    if asset_ids:
        # If this AEZ has multiple parts, add them all; mosaic will merge them.
        for aid in asset_ids:
            images.append(ee.Image(aid))
    else:
        missing.append(aez_no)

if not images:
    raise RuntimeError("No AEZ images found across v4 / v4_local_parts / v4_local")

# ---- Final mosaic across all gathered images ----
# Order: AEZ 1..19; within an AEZ, parts in ascending numeric order.
# ImageCollection.mosaic() draws later images on top of earlier ones.
mosaic_img = ee.ImageCollection(images).mosaic()
m = geemap.Map()
#m.centerObject(farm.limit(1))


url = 'https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}'
m.layout.height = '1000px'
m.add_tile_layer(url, name="Google Map", attribution="Google")
#m.add_basemap(, )
m.addLayerControl()
#m.addLayer(fields_boundary, {}, 'Fields')

palette = [
    '303000', '000000', '000000', '000000', '000000', 'f75cff', 
    '991695', 'e6ab2c', '3bef34', '2baa25', 'e6ef34', 'a1a524', 'eea15e'
]

palette_corrected = [
    '303000', '000000', '000000', '000000', '000000', 'f75cff', 
    '991695', 'e6ab2c', '3bef34', '2baa25', 'e6ef34', 'a1a524', 'eea15e', '29554E'
]

vis_params = {
    'min': 0,
    'max': 12,
    'palette': palette
}

vis_params_corrected = {
    'min': 0,
    'max': 13,
    'palette': palette_corrected
}

vis_params_wri = {
    'min': 0,
    'max': 2,
    'palette': ['ffffff', 'ffffff', 'ff9916']
}
m.addLayer(wri_map, vis_params_wri, 'WRI Grassland Map')
labels = """0. Background
1. Built-up
2. Water (Kharif)
3. Water (Kharif + Rabi)
4. Water (Kharif + Rabi + Zaid)
5. Croplands
6. Tree/Forest
7. Barren Land
8. Single Kharif (Light blue)
9. Single Non-Kharif
10. Double (Dark-Blue)
11. Triple
12. Shrub_Scrub
13. Plantation"""
labels = [i.split(". ")[-1] for i in labels.split("\n")]

palette_ = ['ffffff', 'ffffff', '0000FF']
vis_params_ = {
    'min': 0,
    'max': 2,
    'palette': palette_
}
aez = ee.FeatureCollection("users/mtpictd/agro_eco_regions")

m.addLayer(mosaic_img.select("predicted_label"), vis_params_corrected, 'LULC_corrected')
m.addLayer(original_lulc_img, vis_params, 'LULC_original')
m.addLayer(aez, {}, "AEZ boundaries")

dw = ee.ImageCollection('GOOGLE/DYNAMICWORLD/V1').filterDate('2024-07-01', '2025-07-01').filterBounds(roi)
dw = dw.select('label').mode()

VIS_PALETTE_DW = [
    '000000',  # water
    '991695',  # trees
    'eea15e',  # grass
    'eea15e',  # flooded_vegetation
    'e6ef34',  # crops
    '303000',  # shrub_and_scrub
    '000000',  # built
    'e6ab2c',  # bare
    '000000',  # snow_and_ice
]

m.addLayer(dw, {'min': 0, 'max': 8, 'palette': VIS_PALETTE_DW}, 'Dynamic World (mode)')
# Optional: style by class_label using filters (if needed)
color_map = {
    'Cultivated grassland': 'green',
    'Natural/Semi-natural grassland': 'orange',
    'Other land cover': 'blue'
}

#Map = geemap.Map(center=[0, 0], zoom=2)

# Add each class as a separate styled layer
for label, color in color_map.items():
    styled_fc = fc.filter(ee.Filter.eq('class_labe', label))
    m.addLayer(
        styled_fc.style(**{'color': color, 'pointSize': 5, 'pointShape': 'circle'}),
        {},
        label
    )

#Map.add_legend(title="Class Labels", labels=list(color_map.keys()), colors=list(color_map.values()))
#Map.add_legend(
#    title="Class Labels",
#    legend_dict=color_map  # This takes a dict like {'label': 'color'}
#)
m.addLayer(sampled.filter(
    ee.Filter.And(
        ee.Filter.neq('predicted_label', 6),
        ee.Filter.neq('predicted_label', 7),
        ee.Filter.neq('predicted_label', 12)
    )), {}, 'Misclassified points')
m
```

```python
LABEL_PROP   = 'class_labe'                              # points label property
GRASS_LABEL  = 'Cultivated grassland'  
lulc = mosaic_img.unmask(0)
#lulc = original_lulc_img.unmask(0)
roi = ee.FeatureCollection("users/mtpictd/agro_eco_regions")
# Filter to only the Natural/Semi-natural grassland points
grass_pts = fc.filter(ee.Filter.eq(LABEL_PROP, GRASS_LABEL)).filterBounds(roi)
sampled = lulc.sampleRegions(
    collection = grass_pts,
    properties = [LABEL_PROP],  # optional; keeps the label on the sampled features
    scale      = 10,
    geometries = True,
    tileScale  = 4
)
count_in_7 = sampled.filter(ee.Filter.eq('predicted_label', 7)).size()
count_in_6 = sampled.filter(ee.Filter.eq('predicted_label', 6)).size()
count_in_12 = sampled.filter(ee.Filter.eq('predicted_label', 12)).size()
count_outside_6_7_12 = sampled.filter(
    ee.Filter.And(
        ee.Filter.neq('predicted_label', 6),
        ee.Filter.neq('predicted_label', 7),
        ee.Filter.neq('predicted_label', 12)
    )
).size()
total_grass_points = grass_pts.size()
result = ee.Dictionary({
    'total_grass_points': total_grass_points,
    'in_class_7 (Barren Land)': count_in_7,
    'in_class_6 (Tree/Forest)': count_in_6,
    'in_class_12 (Shrub_Scrub)': count_in_12,
    'sum_in_6_7_12': count_in_6.add(count_in_7).add(count_in_12),
    'percentage_in_6_7_12': count_in_6.add(count_in_7).add(count_in_12).divide(total_grass_points).multiply(100),
    'outside_classes_7_and_6': count_outside_6_7_12
})

# Trigger evaluation on the client
print(result.getInfo())
```
